```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('Heart.csv')
data.head()
```

```python
#Scatterplot
x=data.Age 
y=data.Chol
```

```python
plt.title("ScatterGraph:")
plt.xlabel("Age")
plt.ylabel("Chol")
plt.scatter(x,y)
```

```python
#BarGraph
plt.show()
x=data.Sex
y=data.Ca
```

```python
plt.title("BarGraph:")
plt.xlabel("Sex")
plt.ylabel("Ca")
plt.bar(x,y)
```

```python
#lineGraph
x=data.Sex
y=data.Slope
```

```python
plt.title("LineGraph:")
plt.xlabel("Sex")
plt.ylabel("Slope")
plt.plot(x,y)
```

```python
#BoxPlot
x=data.Chol
```

```python
plt.title("BoxPlot")
plt.ylabel("Chol")
plt.boxplot(x)
```
