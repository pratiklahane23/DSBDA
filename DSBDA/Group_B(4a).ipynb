```python
#Yash Gharate
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('AirQualityodisha.csv')
```

```python
#histogram
plt.figure(figsize=(8,6))
sns.histplot(df['SO2'],bins=10,kde=True,color='green')
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(10,20))
sns.scatterplot(df['SO2'])
sns.scatterplot(df['NO2'])
plt.show()
```

```python
#barplot
plt.figure(figsize=(10,10))
sns.barplot(x='RSPM/PM10',y='PM 2.5',data=df)
```

```python
#boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='SO2',y='NO2',data=df)
```

```python
#distplot
plt.figure(figsize=(8,6))
sns.displot(df['SO2'],bins=10,kde=False,color='yellow')
sns.displot(df['NO2'],bins=10,kde=False,color='PINK')
```

```python
#LINEPLOT
plt.figure(figsize=(10,10))
sns.lineplot(data=df,x='SO2',y='NO2')
```
